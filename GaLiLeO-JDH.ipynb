## Project History

The idea for this project began as a conversation in April 2016 at Stanford University about the points of connection and similar goals of two collaborative efforts that had already been underway for several years: developing an editorial apparatus for Galileo's correspondence based on close reading of the available letters and creating a digital and computational library of the books that Galileo owned. From the perspective of History of Science, Hannah Marcus had worked with a team of scholars to identify thematic overlap in the surviving correspondence related to Galileo. Part of that project also involved careful attention to metadata about the letters, which has resulted in new findings. <cite data-cite="1584927/5R7T42AP"></cite> From the perspective of Digital Humanities, I have been working with colleagues at my institution to develop an analytical framework for digital artifacts that evaluates the role of the digital (things represented by binary) and the computational (the practices enabled through digital representation). <cite data-cite="1584927/LIZJVZTN"></cite> We articulate our approach as a two-way street, in which those artifacts, for example a computational text analysis tool, can be applied to questions and objects of study in the Liberal Arts (Arts, Humanities, Social Science, Physical Sciences). But the tool can also be designed or critiqued with the priorities of those disciplines at the forefront. In this context, my practice of Digital Humanities is concerned with more than the application of computational tools in the study of humanistic expression. I also investigate the ways in which humanistic priorities can guide the development of such tools. With this in mind, Hannah and I explored the extent to which humanistic intervention could reshape the process of digitization, the construction of any underlying data structures for GaLiLeO, the design of the user interface, and the nature of the tools. This article will focus on the data and tools. Given our overlapping goals of learning more about Galileo and his contemporaries through the study of text, we began to design what became GaLiLeO: Galileo's Library and Letters Online.

Naming credit for GaLiLeO goes to Hannah Marcus and Morgan MacLeod after much discussion about what this project would be. We struggled with the identity of the hybrid tools-text nature of the platform. We were not trying to create a digital archive, which would emphasize findability and a range of related material. We could aspire to the William Blake Archive, but we were a small group without funding and broad technical support. <cite data-cite="1584927/H6IM78UH"></cite> We knew that we were not aiming for a digital edition, even though Marchus' team had created a rigorous and extensive scholarly interpretive layer for the correspondence. Using the vague "online" descriptor avoided these problems of identity. We hoped that it would gesture toward actions of aggregation, exploration, and experimentation that is enabled by the web. While bringing together plain, full text of the letters and limited contents from the library would create bridges across otherwise siloed resources, our aim was not to make a claim about a collection but to evaluate materials, methods, and outcomes of analyses of those texts.

Our enthusiasm quickly outpaced resources of time and funding. We composed an ambitious application for a Digital Humanities Advancement Grant through the National Endowment for the Humanities in the United States. <cite data-cite="1584927/B5Q585LU"></cite> The feedback on the failed proposal helped us to scale back our initial tasks to a proof-of-concept for the combined text analysis of the two corpora and their metadata. We maintained the same intellectual priorities outlined in the grant, but worked with materials at hand to focus more on method.

While current use of GaLiLeO is limited to the two principle investigators (and the exploration facilitated by mybinder functionality with this article), it has always been intended for a broader audience of Galileo scholars, interested peers in history and literature, and students. Thanks to financial support from Harvard University's Dean's Competitive Fund for Promising Scholarship, we were able to invite scholars to campus to test this alpha prototype and brainstorm next steps. The resulting September 2018 workshop was an opportunity to evaluate our ground-up approach to humanistic intervention in digitization and computational analysis.

## The Digital Galileo Ecosystem

The primary digital Galileo resource for specialists in the galileo//thek@ (Galileoteca) developed and maintained by the Galileo Museum in Florence, Italy. <cite data-cite="1584927/K7RFDQH5"></cite> The Galileoteca provides access to primary documents, scholarly editions, and secondary scholarship. Like most of Galileo scholarship, the primary documents in the Galileoteca are heavily influenced by the late-19th century National Edition of Galileo's works and correspondence compiled by Antonio Favaro (1847-1922) and his team of collaborators. <cite data-cite="1584927/XYL4T68J"></cite> Documents are organized according to their location in the national edition and digitized text reflects Favaro's editorial interventions, most often in materials written by Galileo, rarely in texts authored by others. The excellent curatorial and editorial team at the Museo Galileo has added other interpretive layers to the digital interface to help visitors find the appropriate documents easily.

In terms of providing historical context, going forward, the other key digital resournce in this ecosystem will be of significant importance: The Galileo Project at Rice University. <cite data-cite="1584927/JEPMCMF5"></cite> Conversations with the specialists at the workshop confirmed that this early Digital Humanities project remains a frequent reference while we conduct our research. Among many other resources on the site, the chronological and bibliographic narratives provided by the project contributors bring to life key events and individuals that are frequent authors or frequently mentioned in the primary source documents.

GaLiLeO seeks to go below these *post hoc* layers and return to the texts to explore new interpretive pathways that have not been predetermined by centuries of scholarship. In the Galileoteca, specialists who know what document they want to find, are studying a chronological period in Galileo's lifetime, or have a thematic or topical research question in mind can use the resources to access related materials. Users can search word forms and lemmas within certain areas of this digital archive or across elevent kinds of materials that also include biographies, indices, and iconography. It is an excellent finding aid for those who already know the specific text they seek. The tools presented here push more toward emphasizing the choices humanists make based on context and allow for the circumstances (the contextual elements) to intervene.

Our approach recognizes the ways in which digital infrastructures instantiate interpretive arguments. Literary and historical inquiry relies on context, often multiple contexts, which is challenging to evaluate in a large corpus if the underlying navigation principle is to quickly demonstrate single documents that match queries. Since Hannah and I were bringing together two corpora that are not typically studied together, our interpretive argument was that the broader print culture surrounding Galileo and his correspondents influenced the letters and likely vice versa. This meant creating textual models and query possibilities that would highlight similarities or differences in a way that gave researchers enough context to both evaluate influence and follow leads suggested by those circumstances.

The availability of data was one of the limiting factors in further development of the project. Even for an individual as famous as Galileo (or perhaps because of it), there is no freely-available set of plain-text files for analysis. The correspondence in the National Edition of Galileo’s works, included over 4,300 letters from the 20 volumes edited by Favaro in the 19th century <cite data-cite="1584927/XYL4T68J"></cite> that have been scanned, at least in their first printing. Many letters were only excerpted in Favaro's edition. Images of manuscripts are housed by different institutions with strict reproduction rules. Material details of papers and books are sparse, as is the information about the development of archives over time. One participant in our workshop spoke about the ideal humanistic digital laboratory offering access to a searchable corpus, allowing for a choice of transcriptions, and images of the originals with an emphasis on physical properties. While many such digital resources exist for an Italian figure like Dante, this goal is still elusive for Galileo studies. 

The field of Galileo studies is moving in directions that still require the attentive philological study of what Galileo and his contemporaries wrote, but with the integration of other perspectives. Some of this work focuses on visual culture, both in terms of diagrams and illustrations but also the arts. Renewed attention has been brought to his marginalia and reading practices. Scholars are also pushing back against claims of exceptionalism to highlight linguistic traditions in which Galileo participated and also the popular topics that he avoided. While GaLiLeO cannot yet address these first shifts, we hope to contribute to the development of scholarship in the last category.

## GaLiLeO and Hermeneutica

One might ask why we did not repurpose a text or metadata analysis tool already in existence for our study. After all, Voyant Tools was one of the primary inspirations for the functions that I wrote for GaLiLeO. <cite data-cite="1584927/2WCMWWUK"></cite> The theoretical framing for Voyant, provided by Rockwell and Sinclair in *Hermeneutica* (2013), is structurally integral to how we teach text analysis at Bowdoin. <cite data-cite="1584927/PVFRKDSE"></cite> In DCS we talk about Abstractions, with an emphasis on models necessarily built into digital artifacts. Even prior to this is the recognition that the digitization process already represents an abstraction from the reality of the physical letter or book to its representation as code. In the best of cases, the digital and computational artifacts, which are the tools of analysis, address more than just an abstraction of a computational process, but also the other humanistic models that Rockwell and Sinclair identify in their exploration of three models in digital humanities: "the theoretical model of the problem or cultural phenomenon; a working model of an instance of the phenomenon based on the first; and a model of the process whereby we will get the second." (163) The tools necessarily draw our attention to certain features while deprioritizing others.

For GaLiLeO, both our analytical goals and the size of our corpus (thousands of documents), made repurposing Voyant challenging. The contextual tools in Voyant draw attention to high-frequency terms moreso than unusual ones. The "Distinctive Words" feature in the "Summary" pane does higlight (literally and figuratively) ways in which authors depart from one another in lexical choices.

Lemmatization and certain editorial practices collapse variants into one form for searching, for instance, all of the letters that talk about the telescope. Yet, this does not help us to distinguish subcategories within the 100 letters returned as results in the Galileoteca. With a term like the related *telescopio*, which has several hundred results, the list needs to be parsed to make sense of the kinds of use of the word. Computation can help.

We quickly realized that we needed a model for asking questions in humanistic fashion rather than a model of the documents related to Galileo. The approach borrows this phrasing from Willard McCarty, distinguishing between an archeological map of a historic site (i.e. the reconstruction of the National Edition of Galileo's works) and the blueprint for a new edifice.<cite data-cite="1584927/4JAXCGB8"></cite> Our model of texts would be forever incomplete and many parts already exist in the ecosystem. A model for inquiry on the other hand would be extensible to the new data sets, for Galileo, but potentially for other authors and collections.

Our first code cells in the prototype made visible the goals and limited implementation of the prototype while obscuring significant interpretive the work. The prepared corpus reflected Galileo's correspondence tagged by Hannah's team and the contents of his library prepared by my team as of July 2017. Metadata for these texts was hand keyed, and we clarified that sometimes authors (including letter writers) are simply unknown. Cleaning the textual data meant interacting with digital representations of these documents' physical materialities. I created and subsequently analyzed two versions of the corpus, one with and one without punctuation, given its inconsistencies in the early modern period. Everything is lowercase. The texts represented a spectrum of quality from a first pass of OCR on scans of 19th-century volumes to OCR results cleaned by comparison to scans of pre-modern printed volumes to documents typed by hand based on the original print or manuscript.

In short, the tools in GaLiLeO embrace the epistemology of *Hermeneutica* <cite data-cite="1584927/PVFRKDSE"></cite> but not the implementation. Since, as Rockwell and Sinclair point out, an "instrument implements a theory of interpreting the phenomenon it was designed to bring into view" (162), we needed a new instrument. Historians are seeking mechanisms to understand tone, reliability (through consistency), corroboration with other viewpoints, and verification. Literary scholars engage with the multiple contexts through which an expression conveys potentially different meanings. For this we needed to bring together not just words and documents, but the overall corpus attributes. Again, as Rockwell and Sinclair so efficiently state in Hermeneutica (2014): "developing tools involves bearing hermeneutical theories." (161) 

This meant that the organization of the texts had to co-evolve with the analytical needs based on the quality and groupings of texts and their associated metadata. The proof of concept identified several areas for future development in terms of customization: how to interact with the specialist editorial layer (or not), choosing a subcorpus of interest, limiting or expanding the number of languages considered, and how to handle Galileo's work. As presented at the workshop, the overall corpus could be subset into specialized document groupings based on type of document and metadata tag categories. Users could include the 167 prefatory letters from 119 titles in Galileo's library. While the "All" option was not available for the prototype demonstration, Hannah and I used it for our subsequent article on crystals, searching for and contextualizing the term across 82 full text files of books in Galileo's library.<cite data-cite="1584927/TUDXHUQF"></cite>

For the sake of simplicity (and time to load the related files), we limited the specialist editorial layer to the documents with "Instruments and Materials" tags. These represented a range of early modern tools for knowledge creation including the telescope and its component pieces, but also candles, compasses, magnets, scales, and vessels for water. The team of specialist readers had also tagged words that were used as descriptors and actions related to experiments and experiences of natural phenomena in service of learning or confirming knowledge. Any text that used one of 905 different terms in this category was included in the subcorpus. 

Overall this created a subcorpus of 2,346 letters for analysis during the workshop. A remaining feature to develop is to included words in titles and metadata as part of the search. Currently the analysis focuses only on the body of documents. Due to an OCR error, the shortest document only had 1 token. The median length of the texts in the subcorpus was 257 words (i.e. 2,178 documents were shorter than this). The corpus represents 9 different languages. A planned feature would allow further subsetting by English, Dutch, French, German, Greek, Italian, Neolatin, Paduan dialect, and Spanish. This would require a team of scholars trained in those languages to help align the tags and clean the OCR.

## GaLiLeO and the Experience of Strangeness

This approach comes with two caveats. The first results from Italian Jesuit Roberto Busa's advocacy for modeling the "conceptual system" of a text's author and the context of that author. (Paraphrased in <cite data-cite="1584927/PVFRKDSE"></cite> p. 152-153). It is a reminder to avoid building tools that apply our own, positivist (i.e. 19th-century), modernist, or post-modernist viewpoints to the detriment of seeing and hearing the system of connections between texts as they were created and experienced in their own historical moment. In other words, I strove to design tools that embraced Italian cultural theorist and historian Carlo Ginzburg's theory of "Making things Strange." <cite data-cite="1584927/YJWVTQER"></cite> Seeing the text differently by decomposing it and visualizing its contents symbolically creates the cognitive distance for rupturing potentially limiting views on our texts. Rather than a remediation of post-hoc lenses with their own epistemological priorities, I want this tool to narrate for me in a way that I, as a 21st-century scholar cannot, the categories, organizing principles, affinities, disparities, patterns, and outliers in the objects that I have collected.

The second caveat builds on this principle and comes from the 17th-century, Galileo's period. It also echoes the concerns raised in *Hermeneutica*. In response to one of Stanley Fish's notable critiques of DH <cite data-cite="1584927/ALTEQGGT"></cite>, Rockwell and Sinclair respond in a way that echoes Galileo's pushback against opponents of his new science. The Voyant creators write that Fish's argument is "like trying to understand railroads by reading books about them instead of looking at models or taking a ride on a Shinkansen" (151) (referencing the Japanese bullet train). Galileo would agree. He offers a more elaborate explanation in the voice of one of the speakers early on the First Day of the *Dialogues on the Two Chief World Systems* (1632): "there are those who possess all of the precepts of da Vinci, but would not know how to paint a stool. Playing an organ is not learned from those who know how to make organs, but from those who know how to play them... painting is learned from continuous drawing and painting." <cite data-cite="1584927/XYL4T68J"></cite> (My translation from VII, 69-70) He goes on, but the point is the same - building an instrument and knowing how to use it are two different things. In the case of an archive informed by previous scholarship, this inherited knowledge is valuable; being able to access the philogical and analytical discoveries of previous scholars through such a database structure is indeed a rich resource. My argument is that we can do more, and Galileo's suggestion that to be a philosopher, or for us to be humanists, we need to act and practice our skills, implies the need for a tool that augments our actions. To continue Galileo's metaphor, in order for our work as humanists to be more than great practice over previous humanists' work, reified in digital form, rather than build the database (the organ in the analogy), we must continually practice our skills of discovery, uncovering, recreating, differentiating, evaluate, contextualizing, and documenting representations and expressions. 

Galileo's lesson about knowing by doing rather than knowing by reading about was deeply specific to the controversies in which he was embroiled over inherited knowledge, particularly that of the classical Greek philosopher Aristotle. His resistance to following the minds of others in philosophy came from experience and experimentation with new tools such as the telescope. Yet, the ethos of skepticism, also expressed in *Hermeneutica*, can also address 21st-century Digital Humanities priorities of de-colonizing archives. In order to know what is interesting enough to search in a corpus, a user needs to arrive with prior knowledge of the values associated with its contents. While not always problematic, systemic exclusion, majority perspectives, and other kinds of bias limit the interpretive pathways of a collection of text structured by technology.  

We wanted to ask contextualized questions as much or more than we wanted to retrieve specific information. For this reason preliminary, contextual analysis occurs prior to the user arriving at the interface. The code is not elegant, but captures quantitative information about each document, the words in the document, and the author. As shown in the figure below, the underlying metadata is also contextualized within the corpus. The results are stored as RDS objects to be loaded into memory when the user chooses a subcorpus for exploration.

The figure above outlines the information created and saved related to each document. Again, there are redundancies and inefficiencies since this is an alpha-prototype. Many of the attributes are recorded with variations: tokenization that includes and excludes punctuation, stylistic similarity with three different feature classes and two distance measures, as well as distinguishing between documents written by Galileo and those written by other authors. The goals were twofold: achieve as many perspectives as possible on the texts and word types as well as explore the strengths or limitations of each perspective in comparison to the others. The method points to both the wunderkammer epistemological approach as well as what Ted Underwood has called perspectival modeling and I have referred to as computational parallax. Creating many of the features also allows for querying by attribute range rather than the specific attribute. For example, by saving the contextual information, a user can identify authors who wrote many documents in the corpus, but were the recipient of none.

Goal is not definition, but experience. Pathways, not interpretations. Question-making and the creation of ways to arrive at answers.

A preliminary topic model was made from each corpus for proof-of-concept purposes. The model was not optimized, so after removing the stop words used for Italian in Voyant Tools, all remaining words (even those occurring in one document only) were considered to infer just 25 topics.
```r
#Get year information
years <- as.numeric()
for (d in 1:length(text_attributes)){
  year <- as.numeric(text_attributes[[d]]$Year)
  years <- c(years, year)
}
all_years <- sort(years, decreasing = F)
years_df <- as.data.frame(table(all_years))
years_df$Color <- "No"
colnames(years_df) <- c("Year", "Count", "Color")
head(years_df)
```

```r
#Partial representation of the function to highlight contextual data about years
choose_doc <- function(text_list, corpus_metadata, full_ID){
  tl_index <- which(Full_IDs == full_ID)
  for (j in 1:length(tl_index)){
    cat("This is 1 of ", text_list[[tl_index[j]]]$AuthorDocs, " documents written by ", text_list[[tl_index[j]]]$Author, "in the corpus.\n")
    cat("The title is: ", corpus_metadata[[tl_index[j]]]$Metadata$Title, "\n")
    cat("\nIt was written or published ", text_list[[tl_index[j]]]$Year, ".\n")
    cat(text_list[[tl_index[j]]]$YearInfo$SameYear, "document(s) was (were) dated or (published) in the same year.\n")
    cat(text_list[[tl_index[j]]]$YearInfo$Before, "document(s) was (were) dated or (published) in the years before it.\n", 
        text_list[[tl_index[j]]]$YearInfo$After, "document(s) was (were) dated or (published) in years after it.\n")
    target_y <- text_list[[tl_index[j]]]$Year
    years_df$Color[which(years_df$Year == target_y)] <- "Yes"
    
    y<-ggplot(years_df, aes(x=as.numeric(Year), y=Count, fill=factor(Color))) +
      geom_bar(stat="identity")+theme_minimal() +
      theme(axis.text.x= element_text(angle = 45)) +
      theme(legend.position = "none") +
      scale_x_continuous(name = "Year", breaks = seq(earliest,latest, by = 5)) +
      labs(title = "Comparison of Selected Text (in Blue) by Year")
    out_dir <- "GaLiLeOResults/"
    year_plot_name <- paste(out_dir, full_ID, "YearComparison.png", sep = "")
    ggsave(filename = year_plot_name, plot = y, device = "png")
    dev.off()
  }
}
```

## Implications and Next Steps

If we continue with the example of Sarrocchi, we can imagine a future instance of GaLiLeO in which we could compare her letters and the text of her epic poem (with which Galileo was familiar) to Galileo's corpus or to the library corpus overall. Her Latin geometrical and theological texts (should they be found) could be distinguished from her letters and text in Italian for a more nuanced comparison of stylistic, thematic, lexical, and contextual similarities or differences.

Beyond this, the plan for GaLiLeO anticipates that any word in the letters or library is:
- a string of alphanumeric characters (ngram)
- a potential editorial tag provided by specialists
- a part of speech (POS)
- a keyword in multiple textual contexts
- a piece of a textual structure (sentence, verse, speaking part, marginalia)
- a piece of multiple semantic networks
- an abstraction (indicative of a dictionary of synonyms and related terms)
- a concept (a collection of instances across time and contexts)
- an expression in time
- a position in vectors of words
- part of metadata
- part of text or paratext
- a material object with a print and/or manuscript history and visual qualities
- an instance of a frequency of usage:
    - In the corpus
    - In works written by Galileo
    - In works written by others

As such, I continue to build tools that allow visual or numerical points of entry for understanding words in their multiple significances in the collection of documents. The GaLiLeO team imagined a dashboard of widgets similar to Voyant, ultimately more user-friendly than the Jupyter notebooks we used to demonstrate the prototype to our colleagues. The day-long workshop was generated more ideas about what a project like this could be. The energy and interest were palpable. We learned about a diverse range of digital user needs that went far beyond the scope of GaLiLeO and we identified immediate and persistent obstacles. We realized immediately at the workshop and later relived when coauthoring our article about Galileo's use of cristallo that the ability to see the words contextualized at multiple levels of addressed both identified points where a collaboration was necessary and the tools facilitated that collaboration. For this literary scholar and digital humanist, I could see the poetic and dramatic points of contact with the texts and narratives that were most familiar to my historian of science collaborator for other attributes. With someone like Sarrocchi, this reveals that the historian might consult with an art historian to better understand the similarity with the painter Cigoli, a social historian or historian of science specializing in the creation and reception of institutions for validating knowledge in early modern Italy, and a literary scholar for the connections between her letters to Galileo and the types of expression found in her epic poem.

The future of the project is still uncertain, pending time and also funding. Here are the general pieces of feedback that we received:
- More tools would be welcome: an API, historical text reuse detection, reporting on avoided topics, named entities, and a “humanistic suggestion tool” that offers models for queries along with rationales.
- More documents would be welcome: marginalia, more library texts, better OCR, and machine readable text based on manuscripts instead of 19th century editorial versions of letters.
- Transmedia linguistic analysis would be a valuable further step.
- GaLiLeO represents an opportunity to bring new perspective on digital projects that are considered “complete” by linking from its terms, people, years, and texts to the lemmario, biographies, chronologies, and bibliographies that have been developed by colleagues in Italy and the U.S. These include the Galileo//thek@, The Galileo Project, and even the authority file of 8,000 names connected to Galilean materials created by the Museo Galileo (http://www.museogalileo.it/esplora/biblioteche/biblioteca/archivionomi.html) 
- The project will benefit from phased roll-outs of the underlying data and metadata, particularly as more texts from the library are added and additional metadata tags are created.
- This is an opportunity to involve undergraduate and graduate students of Italian in an exploration of the history of Italy as well as the ways in which Italian content can influence the development of predominantly Anglophone technology. It could also be a platform to put specialist, informed readers in contact with novice, curious readers.

On the whole, GaLiLeO occupies a connective space across the siloes of robust digital resources for Galileo Galilei, with an opportunity to expand. Ultimately, embedding the project within this ecosystem would benefit the field more than establishing something definitive that must stand alone. Using this tool in conjunction with the lemmatized resources in the Galileo//thek@ or the names in The Galileo Project maintains the experimental, exploratory spirit of the GaLiLeO tools. The Archilet - Reti Epistolari project has already created a model for this with metadata related to literary correspondence in Italy 1600-1800. <cite data-cite="1584927/EPV67BFC"></cite> The computational and quantitative results become a window onto the collection and indication of fruitful pathways, rather than a declaration that foreclose interpretation. By providing a way to enter the corpus without the 19th-century interpretive lens, scholars should be able to benefit nonetheless from that philological and editorial work along previously undefined paths of inquiry.
